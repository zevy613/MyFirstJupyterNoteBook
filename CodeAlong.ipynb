```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#if we started the file path with / it would look at the C drive. basically it would become an absolute path.
df = pd.read_csv('Data/data_science_salary_2021.csv')
df.info()
df.head()
```

```python
df.groupby('job_title_sim')['Avg Salary(K)'].mean().plot(kind='bar')
plt.savefig('lessCoolPlot.png')
```

```python
ax = sns.barplot(data=df, x ='job_title_sim', y= 'Avg Salary(K)')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, ha='right')
plt.savefig('barPlot.png')
```
